```python
#import modules
import pandas as pd
import numpy as np
import os #to deal with files

import seaborn as sns
import matplotlib.pyplot as plt
import librosa #audio file
import librosa.display
from IPython.display import Audio # audio importing
import warnings
warnings.filterwarnings("ignore")
```

```python
##load the datasets
```

```python
# Specify the directory path where your datasets are stored locally
dataset_directory = '/TESS Toronto emotional speech set data'

paths = []
labels = []

# Iterate through files in the specified directory
for dirname, _, filenames in os.walk(dataset_directory):
    for filename in filenames:
        # Append input file paths to the 'paths' list
        paths.append(os.path.join(dirname, filename))
        
        # Extract label from filename
        label = filename.split("_")[0].lower()  # Extract the first part of the filename as the label
        labels.append(label)

print("Database is loaded")
```

```python
#create a data frame
df = pd.DataFrame()
df['speech']= paths 
df['label']= labels
df.head()

```

```python
def waveplot(data, sr, emotion):
    plt.figure(figsize=(10,4))
    plt.title(emotion, size=20)
    librosa.display.waveshow(data, sr=sr)
    plt.show()
    
def spectogram(data, sr, emotion):
    x = librosa.stft(data) #
    xdb = librosa.amplitude_to_db(abs(x))#audio amplitude convert into the log forn Db
    plt.figure(figsize=(11,4))
    plt.title(emotion, size=20)
    librosa.display.specshow(xdb, sr=sr, x_axis='time', y_axis='hz') #hz is hertz and specshow help to display it to waveshow
    plt.colorbar()
```

```python
emotion="fear.wav"
path=df["speech"][df["label"]==emotion][0]
data,sampling_rate=librosa.load(path)
waveplot(data,sampling_rate,emotion)
spectogram(data,sampling_rate,emotion)
Audio(path)
```

```python
#feature extraction
#mfcc is basically transforming the linear frequency and derived from stff into MFCC
```

```python
def extract_mfcc(filename):
    y, sr = librosa.load(filename, duration=3, offset=0.5)
    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)
    return mfcc
```

```python
extract_mfcc(df["speech"][0])
```

```python
X_mfcc = df['speech'].apply(lambda x: extract_mfcc(x))
```

```python
X_mfcc
```

```python
X= [x for x in X_mfcc]
X=np.array(X)
X.shape
```

```python
#input split
X= np.expand_dims(X,-1)
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
y=enc.fit_transform(df[["label"]])
```

```python
y=y.toarray()
```

```python
print(y.shape)

```

#lstm model

```python
#model architecture is well-suited for sequence prediction tasks

from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout
model = Sequential([
    LSTM(30,return_sequences=False,input_shape=(40,1)),#False indicates it only returns output at the last time step
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(7, activation="softmax")#Softmax activation ensures output values represent class probabilities.
    
])
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])#"categorical_crossentropy" loss function, commonly used for multi-class classification.
model.summary()
```

```python
history=model.fit(X,y,validation_split=0.2,epochs=100,batch_size=512,shuffle=True) #used to validated the training history and monitore the model performance
```

```python
epochs=list(range(100))
acc=history.history["accuracy"]
val_acc= history.history["val_accuracy"]

plt.plot(epochs,acc,label="train accuracy")

plt.plot(epochs,val_acc,label="val accuracy")
plt.ylabel("accuracy")
plt.xlabel("epochs")
plt.legend()
plt.show()
```

```python
#visualizing for train accuracy loss
loss=history.history["loss"]
val_loss= history.history["val_loss"]

plt.plot(epochs,loss,label="loss")

plt.plot(epochs,val_loss,label="val loss")
plt.ylabel("loss")
plt.xlabel("epochs")
plt.legend()
plt.show()
```
